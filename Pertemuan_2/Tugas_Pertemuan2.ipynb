# **1. Load Library**

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
````

Keterangan :
1. import pandas as pd
  - Pandas: pandas adalah library Python yang digunakan untuk manipulasi data. Dalam kode ini, pd digunakan sebagai alias untuk pandas, dan Anda akan menggunakannya untuk membuat, membaca, memanipulasi, dan menganalisis DataFrame (struktur data mirip tabel).
2. from sklearn.preprocessing import LabelEncoder, StandardScaler
  - LabelEncoder: Ini digunakan untuk mengonversi kolom data kategori (misalnya, teks seperti "Malignant" atau "Benign") menjadi nilai numerik. Setiap kategori unik diberi label numerik.
  - StandardScaler: Digunakan untuk menstandarisasi fitur numerik sehingga memiliki mean = 0 dan standar deviasi = 1. Ini sering digunakan untuk skala fitur sebelum memasukkannya ke dalam model pembelajaran mesin.
3. from sklearn.model_selection import train_test_split
  - train_test_split: Fungsi ini digunakan untuk membagi dataset menjadi dua bagian, yaitu training set (data yang digunakan untuk melatih model) dan testing set (data yang digunakan untuk menguji performa model). Anda bisa menentukan proporsi data yang dialokasikan untuk masing-masing set, misalnya, 80% untuk training dan 20% untuk testing.

**Rangkuman :**

- Pandas (pd): Memanipulasi data, biasanya dalam bentuk DataFrame.
- LabelEncoder: Mengonversi variabel kategori menjadi numerik.
- StandardScaler: Menstandarisasi data numerik.
- train_test_split: Membagi data menjadi set pelatihan dan pengujian untuk memvalidasi model pembelajaran mesin.

    - id: Biasanya merupakan kolom pengidentifikasi unik yang tidak berkontribusi pada analisis.
    - Unnamed: 32: Kolom ini mungkin merupakan kolom kosong atau berisi nilai yang tidak berguna.
  - axis=1: Parameter ini menunjukkan bahwa yang dihapus adalah kolom. Jika Anda ingin menghapus baris, Anda bisa menggunakan axis=0.

Setelah menjalankan perintah ini, DataFrame df yang baru disimpan kembali ke variabel df dengan dua kolom tersebut dihapus.

**Rangkuman:**

- drop() menghapus kolom yang tidak diperlukan, dalam kasus ini 'id' dan 'Unnamed: 32'.
- axis=1 mengindikasikan penghapusan kolom.

# **5. Encoding**

````python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding

df.head()
````

Keterangan :
1. le = LabelEncoder()
  - LabelEncoder(): Ini adalah bagian dari library sklearn.preprocessing yang digunakan untuk mengonversi nilai kategori menjadi nilai numerik.
  - le adalah objek yang diciptakan dari LabelEncoder(), yang nantinya akan digunakan untuk proses encoding.
2. df['diagnosis'] = le.fit_transform(df['diagnosis'])
  - fit_transform(): Metode ini melakukan dua hal:
    - fit(): Menyesuaikan encoder pada data. Dalam hal ini, LabelEncoder akan memetakan setiap nilai kategori yang ada di kolom 'diagnosis' ke angka. Misalnya, jika kolom diagnosis berisi dua kategori, seperti 'M' (Malignant) dan 'B' (Benign), encoder akan memetakan keduanya ke nilai numerik, seperti 'M' menjadi 1 dan 'B' menjadi 0.
    - transform(): Setelah pemetaan ditentukan, transformasi akan dilakukan untuk mengonversi setiap nilai di kolom 'diagnosis' menjadi angka sesuai pemetaan yang sudah dibuat.
  - df['diagnosis']: Ini menunjuk pada kolom diagnosis dalam DataFrame df, yang di-overwrite dengan hasil transformasi (data yang sudah dikonversi menjadi angka).

Hasilnya, kolom diagnosis yang awalnya berisi nilai kategori seperti 'M' (Malignant) dan 'B' (Benign) akan diubah menjadi nilai numerik, misalnya:

- 1 untuk Malignant
- 0 untuk Benign

**Rangkuman:**

- LabelEncoder: Digunakan untuk mengonversi data kategori menjadi nilai numerik.
- fit_transform(): Melakukan proses fitting dan transformasi secara bersamaan.
- df['diagnosis']: Kolom diagnosis diubah menjadi nilai numerik berdasarkan pemetaan yang dibuat.

# **6. Standarisasi**

````python
# Get numeric columns (excluding 'diagnosis')
numeric_column = df.columns.drop('diagnosis')

# Standarisasi
std = StandardScaler()
df[numeric_column] = std.fit_transform(df[numeric_column])

df.head()
````

Keterangan :
1. Mendapatkan Kolom Numerik (Kecuali 'diagnosis'):

  `numeric_column = df.columns.drop('diagnosis')`

  - df.columns: Mengambil nama semua kolom dalam DataFrame df.
  - .drop('diagnosis'): Menghapus kolom 'diagnosis' dari daftar nama kolom. Hasilnya adalah daftar kolom numerik yang akan diproses.
2. Standarisasi Data:

  ```
  std = StandardScaler()
  df[numeric_column] = std.fit_transform(df[numeric_column])
  ```
  - std = StandardScaler(): Membuat objek StandardScaler dari sklearn.preprocessing. StandardScaler digunakan untuk standarisasi fitur dengan mengubah distribusi data menjadi distribusi normal dengan rata-rata 0 dan deviasi standar 1.
  - std.fit_transform(df[numeric_column]):
    - fit(): Menghitung statistik yang diperlukan (rata-rata dan deviasi standar) pada kolom numerik.
    - transform(): Menggunakan statistik yang dihitung untuk mengubah nilai-nilai kolom numerik menjadi data yang terstandarisasi.
  - df[numeric_column] = ...: Menyimpan hasil standarisasi kembali ke DataFrame df, menggantikan kolom numerik asli dengan nilai-nilai yang telah terstandarisasi.

  
  **Ringkasan**:

- Kode ini menghapus kolom 'diagnosis' dari daftar kolom yang akan diproses.
- Menggunakan StandardScaler untuk standarisasi kolom numerik (mengubah skala fitur menjadi distribusi dengan rata-rata 0 dan deviasi standar 1).
- Menyimpan hasil standarisasi ke dalam DataFrame dan menampilkan lima baris pertama untuk memeriksa hasilnya.

# **7. Split Data**

````python
# Split data training dan testing
# Rasio yang akan kita gunakan adalah 8:2 (Training : Testing)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
````

Keterangan :

1. Membagi Data Menjadi Data Pelatihan dan Pengujian:

  ```
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])
  ```

  - train_test_split(): Fungsi dari sklearn.model_selection yang membagi data menjadi dua subset: pelatihan dan pengujian.
  - df: DataFrame yang berisi data yang akan dibagi.
  - test_size=0.2: Menentukan proporsi data yang akan digunakan sebagai data pengujian (20% dari total data). Sisanya (80%) digunakan sebagai data pelatihan.
  - random_state=0: Menetapkan nilai acak untuk memastikan pembagian data yang konsisten setiap kali kode dijalankan. Ini memungkinkan reproduktifitas hasil.
  - stratify=df['diagnosis']: Memastikan bahwa proporsi label dalam data pelatihan dan pengujian sama dengan proporsi label dalam data asli. Dengan kata lain, stratifikasi menjaga distribusi label yang seimbang di kedua subset.
2. Memeriksa Distribusi Label:

  ```
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
  ```

  - df.diagnosis.value_counts(): Menghitung jumlah kemunculan masing-masing label dalam data asli. Ini memberikan distribusi label di data sebelum dibagi.
  - df_train.diagnosis.value_counts(): Menghitung jumlah kemunculan masing-masing label dalam data pelatihan. Ini menunjukkan distribusi label di data pelatihan.
  - df_test.diagnosis.value_counts(): Menghitung jumlah kemunculan masing-masing label dalam data pengujian. Ini menunjukkan distribusi label di data pengujian.
  - print(): Menampilkan distribusi label dalam data asli, pelatihan, dan pengujian untuk memastikan bahwa pembagian data mempertahankan distribusi label yang serupa.

**Ringkasan**:

- Kode ini membagi data menjadi set pelatihan dan pengujian dengan rasio 80:20.
- Menggunakan stratifikasi untuk memastikan bahwa distribusi label di data pelatihan dan pengujian mirip dengan data asli.
- Memeriksa distribusi label di setiap set untuk memastikan pembagian data dilakukan dengan benar dan distribusi label tetap konsisten.
